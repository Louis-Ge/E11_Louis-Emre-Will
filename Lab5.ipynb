```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import time
import random
```

```python
header = ['Time','0.3','0.5','1.0','2.5','Temperature','Pressure','Humidity']
inside = pd.read_csv('aq_temp_inside.csv', header=[0,1,2,3,4,7,8,9])
outside = pd.read_csv('aq_temp_outside.csv', header=[0,1,2,3,4,7,8,9])

for col in header[1:]:
    ax = plt.axes()
    ax.plot(inside[col][1:], label='Inside')
    ax.plot(outside[col][1:], label='Outside')

    ax.set(ylabel=col, xlabel='Time (s)')

    plt.legend()
    plt.show()
```

```python
for col in header[1:]:
    ax = plt.axes()
    ax.hist(inside[col][1:], label='Inside')
    ax.hist(outside[col][1:], label='Outside')

    ax.set(xlabel=col, ylabel='Count')

    plt.legend()
    plt.show()
```

```python
for col in header[1:]:
    dfi = 1
    means = []
    print(f'┌ \033[1m{col}:\033[0m')
    for df in [inside, outside]:
        dfi += 1
        mean = np.mean(df[col][1:], axis=0)[-1]
        std = np.std(df[col][1:], axis=0)[-1]
        #std_mean = std / np.sqrt(len(df[col][1:]))
        means.append(mean)
        print(f'|    {['Inside','Outside'][dfi % 2]}:')
        print(f'|      Mean: {mean}\n|      Standard deviation: {std}')
    
    std_mean = np.std(means) / np.sqrt(2)
    print(f'|\033[0;35mStandard deviation on the mean: {std_mean}\033[0m')
    if abs(means[0]-means[1]) > 2.8*std_mean:
        print(f'└\033[0;31mStatistically different?: Yes\033[0m')
    else:
        print(f'└\033[0;32mStatistically different?: No\033[0m')
    
```
