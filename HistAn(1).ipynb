```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

insidedata = pd.read_csv("https://raw.githubusercontent.com/Louis-Ge/E11_Louis-Emre-Will/refs/heads/main/data/aq_temp_inside.csv")
outsidedata = pd.read_csv("https://raw.githubusercontent.com/Louis-Ge/E11_Louis-Emre-Will/refs/heads/main/data/aq_temp_outside.csv")

outsidedata = outsidedata[60:]
insidedata = insidedata[60:]

fig, axs = plt.subplots(7)
fig.set_figheight(25)
fig.set_figwidth(12)


axs[0].hist(insidedata["Temperature"])
axs[0].hist(outsidedata["Temperature"])
axs[0].set_ylabel("Occurences")
axs[0].set_xlabel("Temperature")

axs[1].hist(insidedata["Pressure"])
axs[1].hist(outsidedata["Pressure"])
axs[1].set_ylabel("Occurences")
axs[1].set_xlabel("Pressure")

axs[2].hist(insidedata["Humidity"])
axs[2].hist(outsidedata["Humidity"])
axs[2].set_ylabel("Occurences")
axs[2].set_xlabel("Humidity")

axs[3].hist(insidedata["0.3"])
axs[3].hist(outsidedata["0.3"])
axs[3].set_ylabel("Occurences")
axs[3].set_xlabel("0.3")

axs[4].hist(insidedata["0.5"])
axs[4].hist(outsidedata["0.5"])
axs[4].set_ylabel("Occurences")
axs[4].set_xlabel("0.5")

axs[5].hist(insidedata["1.0"])
axs[5].hist(outsidedata["1.0"])
axs[5].set_ylabel("Occurences")
axs[5].set_xlabel("1.0")

axs[6].hist(insidedata["2.5"])
axs[6].hist(outsidedata["2.5"])
axs[6].set_ylabel("Occurences")
axs[6].set_xlabel("2.5")


plt.legend()
```

Comparing means/standard deviations:

```python
std_T_In=np.std(insidedata["Temperature"])
std_T_Out=np.std(outsidedata["Temperature"])
std_P_In=np.std(insidedata["Pressure"])
std_P_Out=np.std(outsidedata["Pressure"])
std_H_In=np.std(insidedata["Humidity"])
std_H_Out=np.std(outsidedata["Humidity"])
std_3_In=np.std(insidedata["0.3"])
std_3_Out=np.std(outsidedata["0.3"])
std_5_In=np.std(insidedata["0.5"])
std_5_Out=np.std(outsidedata["0.5"])
std_1_In=np.std(insidedata["1.0"])
std_1_Out=np.std(outsidedata["1.0"])
std_2_In=np.std(insidedata["2.5"])
std_2_Out=np.std(outsidedata["2.5"])
meanit=np.mean(insidedata["Temperature"])
meanot=np.mean(outsidedata["Temperature"])
meanip=np.mean(insidedata["Pressure"])
meanop=np.mean(outsidedata["Pressure"])
meanih=np.mean(insidedata["Humidity"])
meanoh=np.mean(outsidedata["Humidity"])
meani3=np.mean(insidedata["0.3"])
meano3=np.mean(outsidedata["0.3"])
meani5=np.mean(insidedata["0.5"])
meano5=np.mean(outsidedata["0.5"])
meani1=np.mean(insidedata["1.0"])
meano1=np.mean(outsidedata["1.0"])
meani2=np.mean(insidedata["2.5"])
meano2=np.mean(outsidedata["2.5"])
nroot=np.sqrt(500)
mstd_T_In=std_T_In/np.sqrt(500)
mstd_T_Out=std_T_Out/np.sqrt(500)
mstd_P_In=std_P_In/np.sqrt(500)
mstd_P_Out=std_P_Out/np.sqrt(500)
mstd_H_In=std_H_In/np.sqrt(500)
mstd_H_Out=std_H_Out/np.sqrt(500)
mstd_3_In=std_3_In/np.sqrt(500)
mstd_3_Out=std_3_Out/np.sqrt(500)
mstd_5_In=std_5_In/np.sqrt(500)
mstd_5_Out=std_5_Out/np.sqrt(500)
mstd_1_In=std_1_In/np.sqrt(500)
mstd_1_Out=std_1_Out/np.sqrt(500)
mstd_2_In=std_2_In/np.sqrt(500)
mstd_2_Out=std_2_Out/np.sqrt(500)


print('Temperature:')

print('insidestd:'), print(std_T_In)
print('outsidestd:'), print(std_T_Out)
print ('insidemean:'), print(meanit)
print ('outsidemean:'), print (meanot)
print ('meanuncertaintyoutside'), print (mstd_T_In)
print ('meanuncertaintyinside'), print (mstd_T_Out)
print('Pressure:')

print('insidestd:'), print(std_P_In)
print('outsidestd:'), print(std_P_Out)
print ('insidemean:'), print(meanip)
print ('outsidemean:'), print (meanop)
print ('meanuncertaintyoutside'), print (mstd_P_In)
print ('meanuncertaintyinside'), print (mstd_P_Out)
print('Humidity:')

print('insidestd:'), print(std_H_In)
print('outsidestd:'), print(std_H_Out)
print ('insidemean:'), print(meanih)
print ('outsidemean:'), print (meanoh)
print ('meanuncertaintyoutside'), print (mstd_H_In)
print ('meanuncertaintyinside'), print (mstd_H_Out)
print('Air Quality (0.3):')

print('insidestd:'), print(std_3_In)
print('outsidestd:'), print(std_3_Out)
print ('insidemean:'), print(meani3)
print ('outsidemean:'), print (meano3)
print ('meanuncertaintyoutside'), print (mstd_3_In)
print ('meanuncertaintyinside'), print (mstd_3_Out)
print('Air Quality (0.5):')

print('insidestd:'), print(std_5_In)
print('outsidestd:'), print(std_5_Out)
print ('insidemean:'), print(meani5)
print ('outsidemean:'), print (meano5)
print ('meanuncertaintyoutside'), print (mstd_5_In)
print ('meanuncertaintyinside'), print (mstd_5_Out)
print('Air Quality (1.0):')

print('insidestd:'), print(std_1_In)
print('outsidestd:'), print(std_1_Out)
print ('insidemean:'), print(meani1)
print ('outsidemean:'), print (meano1)
print ('meanuncertaintyoutside'), print (mstd_1_In)
print ('meanuncertaintyinside'), print (mstd_1_Out)
print('Air Quality (2.5):')

print('insidestd:'), print(std_2_In)
print('outsidestd:'), print(std_2_Out)
print ('insidemean:'), print(meani2)
print ('outsidemean:'), print (meano2)
print ('meanuncertaintyoutside'), print (mstd_2_In)
print ('meanuncertaintyinside'), print (mstd_2_Out)
```

In general, the inside temperature measured higher, the pressure was around the same inside and outside, the humidity outside was higher, and more particles were detected of each size outside.

```python
plt.scatter((insidedata["Temperature"]), (insidedata["Humidity"]))
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")
plt.title("Temperature vs Humidity")
plt.show()
```

The hidden variable in this plot is time because both humidity and temperature were taken as functions of time through the sensor, if it weren't present then there would be no easy way to relate the two. This plot seems to show a linear regression overall where temperature and humidity are  related in negative proportionallity, as temperature increases the humidity decreases. This could be quantified by finding the slope of the line of best fit represented by the scatterplot relating the two.
